```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
import math as m
import ctypes as ct

lib = ct.CDLL('./lib.so')
test_sorting = lib.test_sorting
test_sorting.restype = ct.c_double
MIN_SIZE = ct.c_int.in_dll(lib, 'MIN_SIZE')
```

### Сортировка слиянием

```python
def plot_min_size(func, arr_sizes, min_sizes):
    def calc_plot(arr_size):
        x, y = np.array([]), np.array([])
        for msize in min_sizes:
            MIN_SIZE.value = msize
            x = np.append(x, msize)
            time = test_sorting(arr_size, func)
            y = np.append(y, time)
        return x, y

    fig, ax = plt.subplots(figsize=(10,6))
    ax.grid()
    ax.set(xlabel='min_size, n', ylabel='time, s')

    def plot(arr_size):
        x, y = calc_plot(arr_size)
        ax.plot(x, y, label=str(arr_size))
        ax.legend()

    for arr_size in arr_sizes:
        plot(arr_size)

f = lib.merge_sort_parallel_test
lib.omp_set_num_threads(6)
plot_min_size(f, [1000000], range(5000, 1000000, 10000))
lib.omp_set_num_threads(2)
plot_min_size(f, [1000000], range(1000, 1000000, 10000))
```

```python
def plot_acc_eff(func, arr_sizes, nproc):
    def calc_plot(arr_size):
        x, y = np.array([]), np.array([])
        for n in range(1, nproc + 1):
            lib.omp_set_num_threads(n)
            x = np.append(x, n)
            time = 100
            for i in range(4):
                time = min(time, test_sorting(int(arr_size), func))
            y = np.append(y, time)
        return x, y

    def plot():
        for arr_size in arr_sizes:
            x, y = calc_plot(arr_size)
            y = y[0] / y
            ax1.plot(x, y, label='array size ' + str(arr_size))
            ax1.legend()
            y = y / x
            ax2.plot(x, y, label='array size ' + str(arr_size))
            ax2.legend()

    fig, ax1 = plt.subplots(figsize=(10,6))
    fig, ax2 = plt.subplots(figsize=(10,6))
    ax1.grid()
    ax2.grid()
    ax1.set(xlabel='threads number', ylabel='acceleration', title='Acceleration')
    ax2.set(xlabel='threads number', ylabel='efficiency', title='Efficiency')
    plot()

plot_acc_eff(lib.merge_sort_parallel, [1e6, 5e6, 1e7], 12)
```

```python
def plot_time(arsizes):
    def calc_time(func):
        x, y = np.array([]), np.array([])
        for size in arsizes:
            x = np.append(x, size)
            y = np.append(y, test_sorting(int(size), func))
        return x, y

    fig, ax = plt.subplots(figsize=(10,6))
    ax.grid()
    ax.set(xlabel='array size, n', ylabel='time, s')

    x, y = calc_time(lib.merge_sort_parallel)
    ax.plot(x, y, label='merge_sort_parallel')
    x, y = calc_time(lib.quick_sort_parallel)
    ax.plot(x, y, label='quick_sort_parallel')
    ax.legend()

plot_time(range(1000000, 10000000, 1000000))
```
